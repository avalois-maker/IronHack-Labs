```python
import os
from openai import OpenAI
import json
from collections import Counter
import time
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

# Set default model
MODEL = "gpt-4o-mini"

print("✅ Setup complete! OpenAI client initialized.")

# Define a safe `call_openai` helper so the notebook won't error.
# If you have an API key it will call the real API; otherwise this
# fallback mock returns simple sentiment labels for testing.
try:
	def call_openai(prompt, system_message="You are a helpful assistant.", temperature=0.7, max_tokens=None, seed=None):
		# If the OpenAI client is available and API key present, call the API
		if 'client' in globals() and os.environ.get("OPENAI_API_KEY"):
			params = {
				"model": MODEL,
				"messages": [
					{"role": "system", "content": system_message},
					{"role": "user", "content": prompt}
				],
				"temperature": temperature
			}
			if max_tokens is not None:
				params["max_tokens"] = max_tokens
			if seed is not None:
				params["seed"] = seed
			resp = client.chat.completions.create(**params)
			return resp.choices[0].message.content
		# Fallback mock for local testing without API key
		text = prompt.lower()
		if any(w in text for w in ["love", "great", "excellent", "amazing"]):
			return "Positive"
		if any(w in text for w in ["hate", "terrible", "awful"]):
			return "Negative"
		return "Neutral"
except Exception:
	# Ensure notebook doesn't crash if something unexpected happens during definition
	def call_openai(prompt, **_):
		text = prompt.lower()
		if any(w in text for w in ["love", "great"]):
			return "Positive"
		if any(w in text for w in ["hate", "terrible"]):
			return "Negative"
		return "Neutral"

print("✅ call_openai helper defined (real API if configured, otherwise mocked).")
```

```python
# My V4 Sentiment Prompt
sentiment_prompt_v4 = """
"Classify this customer message: ""I dislike this product! It does't work most of the time and I want to return it.""
Example Output: ""Very Unsatisfied"""

# Run the prompt multiple times to compare results
num_runs = 15
results = []
for i in range(num_runs):
	result = call_openai(sentiment_prompt_v4)
	print(f"Run #{i+1} result:")
	print(result)
	print("-"*40)
	results.append(result)
```

```python
# My V4 Product Description Prompt
product_prompt_v4 = """
Create a product description for a laptop that costs $500.
Example Output should include:
1. Target audience as students going backt to school in September
2. Length no more than 50 words
3. Example output: "Powerful yet affordable, this laptop will help student's succeed at whatever challenges the semester can throw at them"""

# Run the prompt multiple times to compare results
num_runs = 15
results = []
for i in range(num_runs):
	result = call_openai(product_prompt_v4)
	print(f"Run #{i+1} result:")
	print(result)
	print("-"*40)
	results.append(result)
```
