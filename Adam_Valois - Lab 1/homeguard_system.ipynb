###Adam Valois_this project is to build a home security system
Security Alerts:
- Motion detected when house is in "away" mode
- Door opened when house is in "away" mode
- Multiple sensors triggered simultaneously (possible break-in)
Safety Alerts:
- Temperature below 35°F (frozen pipe risk)
- Temperature above 95°F (equipment failure)
- Smoke detected (fire risk)
Comfort Notifications:
- Temperature outside comfort range (65-75°F) when home is in "home" mode
- Unusual patterns (like a door left open for >5 minutes)


```python
#Steptest imports
import datetime
import random
ct = datetime.datetime.now()
print("current time:", ct)
temp = random.randrange(-100,100,1)
print("current temperature:", temp)
```

```python
#Step 1: Define Dictionaries
temp_sensor = {"temperature": temp}
smoke_sensor = {"safe": False,"smoke detected": True}
door_sensor = {"open": False,"closed": True}
house_mode = {"home": True,"away": False}
temp_log = {"temperature": temp_sensor["temperature"],"time": ct}
smoke_log = {"smoke detected": smoke_sensor["smoke detected"],"time": ct}
door_log = {"door status": door_sensor["open"],"time": ct}
```

```python
#Step 2: Creating Data Structures
def initialize_system():
    print("Initializing HomeGuard System...")
    print("System initialized at:", ct)
    print("Current temperature:", temp_sensor["temperature"])
    print("Smoke detected:", smoke_sensor["safe"])
    print("Door closed:", door_sensor["closed"])
    print("Home:", house_mode["home"])
initialize_system()
```

```python
#Step 3: Implementing Functions
def check_temperature():
    if temp_sensor["temperature"] < 35:
        print("Warning: Frozen Pipe Risk!")
        temp_log.update({"temperature": temp_sensor["temperature"], "time": ct})
    elif temp_sensor["temperature"] > 95:
        print("Warning: Temperature is above normal, equipment may fail!")
        temp_log.update({"temperature": temp_sensor["temperature"], "time": ct})
    else:
        print("Temperature is normal.")
def check_smoke():
    if smoke_sensor["smoke detected"]:
        print("Warning: Smoke detected! Evacuate immediately!")
        smoke_log.update({"smoke detected": smoke_sensor["smoke detected"], "time": ct})
    else:
        print("No smoke detected.")
def check_door():
    if door_sensor["open"]:
        print("Warning: Door is open! Security risk!")
        door_log.update({"door status": door_sensor["open"], "time": ct})
    else:
        print("Door is closed.")
        door_log.update({"door status": door_sensor["closed"], "time": ct})
def comfort_mode():
    if house_mode["home"]:
        if temp_sensor["temperature"] < 65 or temp_sensor["temperature"] > 75:
            print("Warning: Temperature is not comfortable. Adjusting thermostat.")
        print("House is comfortable. Enjoy your stay!")
    else:
        print("You are away from home. Security mode is active.")
```

```python
def system_status():
    print("System Status:")
    check_temperature()
    check_smoke()
    check_door()
    comfort_mode()
system_status()
print("Temperature Log:", temp_log)
print("Smoke Log:", smoke_log)
print("Door Log:", door_log)
print("System check complete at:", ct)
```
